#  Unemployee Data analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('unemployment dataframe.csv')
df
```

```python
df=df.set_index('Country Name')
```

# Top 10 Least unemployed

```python
df['2021'].sort_values().head(10).plot.bar()
df['2021'].sort_values().head(10)
```

# Top 10 Most unemployed

```python
df['2021'].sort_values(ascending=False).head(10).plot.bar()
df['2021'].sort_values(ascending=False).head(10)
```

## the most, the least

```python
print("the most unemployee")
print(df['2021'].sort_values(ascending=False).head(1).to_string())
print("the least unemployee")
print(df['2021'].sort_values().head(1).to_string())
```
